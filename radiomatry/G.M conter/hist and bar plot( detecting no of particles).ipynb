```python
import matplotlib.pyplot as plt
import numpy as np

plt.style.use('_mpl-gallery')

# make data:
bg=23.55
x = np.arange(1, 17, 1)
y = np.array([189.33, 149.67, 115, 592.67, 5796, 49.33, 129.33, 864.3, 237, 633.67, 1454, 2250.33, 327, 1121.6, 28, 24])
y=y-bg
# plot
fig, ax = plt.subplots(figsize=(12, 8))  # Adjusting figure size

ax.bar(x, y, width=0.8, edgecolor="white", linewidth=0.7)

# Adjusting axis limits
ax.set(xlim=(0, 17), xticks=np.arange(1, 18, 1),
       ylim=(0, max(y) + 500), yticks=np.arange(0, max(y) + 500, 500))

plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns  # Import Seaborn

# Constants
bg = 23.55
x = np.arange(1, 17, 1)
y = np.array([189.33, 149.67, 115, 592.67, 5796, 49.33, 129.33, 864.3, 237, 633.67, 1454, 2250.33, 327, 1121.6, 28, 24])

# Subtract bg from each element in y
y = y - bg

# Create a DataFrame
data = pd.DataFrame({
    'x': x,
    'y': y
})

# Plot bar plot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(x='x', y='y', data=data)

plt.xlabel('S.N')
plt.ylabel('no of particles detected')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
plt.hist(y, bins=10, edgecolor='k', alpha=0.7)  # Adjust bins as needed
plt.title('Histogram of Adjusted y Values')
plt.xlabel('Adjusted y Values')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Constants
bg = 23.55
x = np.arange(1, 17, 1)
y = np.array([189.33, 149.67, 115, 592.67, 5796, 49.33, 129.33, 864.3, 237, 633.67, 1454, 2250.33, 327, 1121.6, 28, 24])

# Subtract bg from each element in y
y = y - bg

# Plot histogram
plt.figure(figsize=(10, 6))
plt.hist(y, bins=10, edgecolor='black', alpha=0.7)

plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram of Particle Counts')
plt.grid(True)
plt.show()
```
